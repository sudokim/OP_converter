```python
from converter import *
from math import nan
from tqdm import tqdm
for row in tqdm(df.to_dict(orient='records')):
    try:
        result = None
        index = row['Index']
        pc = PostfixConverter()

        # if 'GEN_POSSIBLE_LIST' in row[0]:
        #     continue
        if row['OP Equation'] == nan or row['Answer'] == nan:
            continue
        print(f'Solving {index}...', end='', flush=True)

        if isinstance(row['OP Equation'], float) or isinstance(row['Answer'], float):
            continue
            
        if row['OP Equation'].startswith('ans ='):
            exec(row['OP Equation'])
            result = ans
        else:
            result, code = pc.convert(row['OP Equation'])
            # print(f"Result: {result} Answer: {row[2]}")

        assert str(result) == str(row['Answer']), 'Wrong Answer!'
        print(f'\r', end='', flush=True)

        del pc

    except Exception as e:
        # print(pc.code_string)
        print(row, sep=' ')
        print(f"** Error: {e}")
        if result is not None:
            print(f"** Result: {result}")
        print("\n\n")
        traceback.format_exc()
        # raise RuntimeError from e
```
